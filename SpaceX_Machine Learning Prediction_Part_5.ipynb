This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

## TASK  2


Standardize the data in <code>X</code> then reassign it to the variable  <code>X</code> using the transform provided below.


```python
# students get this 
transform = preprocessing.StandardScaler()
# Importer le StandardScaler
from sklearn.preprocessing import StandardScaler

# Créer le transformateur
transform = StandardScaler()

# Appliquer la transformation et réassigner à X
X = transform.fit_transform(X)

# Vérification de la forme et aperçu
print(X.shape)
print(X[:5])  # affiche les 5 premières lignes
```

## TASK  3


Use the function train_test_split to split the data X and Y into training and test data. Set the parameter test_size to  0.2 and random_state to 2. The training data and test data should be assigned to the following labels.


<code>X_train, X_test, Y_train, Y_test</code>


```python
from sklearn.model_selection import train_test_split

# Split les données en 80% training et 20% test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

# Vérification
print("Taille X_train:", X_train.shape)
print("Taille X_test:", X_test.shape)
print("Taille Y_train:", Y_train.shape)
print("Taille Y_test:", Y_test.shape)

# On vérifie que Y_test contient bien 18 échantillons
print("Nombre d'échantillons dans Y_test:", Y_test.shape[0])
```

Create a support vector machine object then  create a  <code>GridSearchCV</code> object  <code>svm_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

# Définition de la grille d’hyperparamètres
parameters = {
    'kernel': ('linear', 'rbf', 'poly', 'sigmoid'),
    'C': np.logspace(-3, 3, 5),
    'gamma': np.logspace(-3, 3, 5)
}

# Modèle SVM
svm = SVC()

# GridSearchCV avec validation croisée 10 plis
svm_cv = GridSearchCV(svm, parameters, cv=10)

# Entraînement sur les données
svm_cv.fit(X_train, Y_train)

# Affichage des meilleurs paramètres et précision validation
print("Tuned hyperparameters :", svm_cv.best_params_)
print("Accuracy (validation) :", svm_cv.best_score_)
```

## TASK  8


Create a decision tree classifier object then  create a  <code>GridSearchCV</code> object  <code>tree_cv</code> with cv = 10.  Fit the object to find the best parameters from the dictionary <code>parameters</code>.


```python
parameters = {'criterion': ['gini', 'entropy'],
     'splitter': ['best', 'random'],
     'max_depth': [2*n for n in range(1,10)],
     'max_features': ['auto', 'sqrt'],
     'min_samples_leaf': [1, 2, 4],
     'min_samples_split': [2, 5, 10]}

tree = DecisionTreeClassifier()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Grille d'hyperparamètres
parameters = {
    'criterion': ['gini', 'entropy'],          # 'entropy' peut émettre un avertissement (alias de log_loss)
    'splitter': ['best', 'random'],
    'max_depth': [2*n for n in range(1, 10)],  # 2,4,...,18
    'max_features': [None, 'sqrt', 'log2'],    # 'auto' n'est pas valide pour DecisionTreeClassifier
    'min_samples_leaf': [1, 2, 4],
    'min_samples_split': [2, 5, 10],
}

# Modèle
tree = DecisionTreeClassifier(random_state=2)

# GridSearchCV (cv=10)
tree_cv = GridSearchCV(tree, parameters, cv=10)

# Entraînement
tree_cv.fit(X_train, Y_train)

# Résultats validation
print("Meilleurs hyperparamètres :", tree_cv.best_params_)
print("Accuracy (validation) :", tree_cv.best_score_)

# Évaluation sur le test set
yhat_tree = tree_cv.predict(X_test)
print("Accuracy sur test data (Decision Tree):", accuracy_score(Y_test, yhat_tree))

# Matrice de confusion
plot_confusion_matrix(Y_test, yhat_tree)
```

```python
print("tuned hpyerparameters :(best parameters) ",tree_cv.best_params_)
print("accuracy :",tree_cv.best_score_)
```
